```python
from sklearn.cluster import KMeans
import pandas as pd
import numpy as np

import warnings
warnings.filterwarnings('ignore')
```

```python
data=pd.read_csv(r"\Users\Slmss\Documents\GitHub\Data_Science-_Dummy\kmn.csv")
```

```python
data.head()
```

```python
df=data.iloc[:,4:]
df.head()
```

```python
from sklearn.preprocessing import StandardScaler

mn=StandardScaler()
X=mn.fit_transform(df)
X
```

```python
from sklearn.decomposition import PCA

pca=PCA(2)
pca.fit(X)
pc=pca.transform(X)
```

```python
inter=[]
for i in range(1,11):
    kmeans=KMeans(i+1)
    kmeans.fit(pc)
    inter.append(kmeans.inertia_)
    
```

```python
import matplotlib.pyplot as plt
# plt.xlim(0,20)
plt.plot(range(1,11),inter)
```

```python
pd.DataFrame(kmeans.cluster_centers_)
```

```python
kmeans=KMeans(6)
clusters=pd.DataFrame(kmeans.fit_predict(pc),columns=['Clusters'])
fdf=pd.concat((data[['Name']],clusters),axis=1,join='outer')
fdf.sort_values('Clusters').reset_index().drop('index',axis=1)
```

```python
kmeans.cluster_centers_[1,:]
```

```python
from sklearn.manifold import TSNE
tsne=TSNE()
scores=pd.DataFrame(tsne.fit_transform(df),columns=['a','b'])

plt.figure(figsize=(20,20),dpi=1024)
plt.subplot(1,2,1)
for i in range(len(fdf.Name)):
    plt.annotate(fdf.Name[i], (scores.a[i],scores.b[i]),size=10)
    scatter=plt.scatter(scores.a,scores.b,c=fdf.Clusters,alpha=0.9)
    plt.legend(*scatter.legend_elements(),loc="upper left", title="Classes")

for i in range(0,6):
    plt.annotate(str(i), (kmeans.cluster_centers_[i,:]),size=10)
    scatter=plt.scatter(kmeans.cluster_centers_[:,0],kmeans.cluster_centers_[:,1],c=fdf.Clusters.unique())
#     plt.legend(*scatter.legend_elements())

# plt.subplot(1,2,2)
# scatter=plt.scatter(kmeans.cluster_centers_[:,0],kmeans.cluster_centers_[:,1],c=[i for i in range(0,6)])
# plt.legend(*scatter.legend_elements())

plt.show()
```

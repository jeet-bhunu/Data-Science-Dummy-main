```python
#Import Library
from numpy import array
import numpy as np
import pandas as pd
from sklearn.model_selection import KFold
```

```python
# data sample
data = pd.read_csv("https://raw.githubusercontent.com/slmsshk/DataSet/main/fmri.csv")
data = np.asarray(data)
```

```python
# prepare cross validation
kfold = KFold(11)
# enumerate splits
split=0
for train, test in kfold.split(data):
    split+=1
    print('train: %s, test: %s split: %i' % (data[train], data[test], split))
    print()
```

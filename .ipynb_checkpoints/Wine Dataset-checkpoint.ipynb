```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

import sklearn.datasets as dt

import warnings
warnings.filterwarnings('ignore')
```

```python
X,y=dt.load_wine(return_X_y=True,as_frame=True)
X.head()
```

```python
from sklearn.manifold import TSNE
from sklearn.decomposition import PCA
```

```python
pca2=PCA(2)
tsne=TSNE()

viz=pca2.fit_transform(X)

plt.figure(figsize=(15,5))
plt.subplot(1,2,1)
plt.title('PCA')
scp=plt.scatter(pd.DataFrame(viz)[0],pd.DataFrame(viz)[1],c=y,alpha=0.5)
plt.legend(*scp.legend_elements(),loc="upper left", title="Classes")


plt.subplot(1,2,2)
plt.title('TSNE')
viz=tsne.fit_transform(X)
scp2=plt.scatter(pd.DataFrame(viz)[0],pd.DataFrame(viz)[1],c=y,alpha=0.5)
plt.legend(*scp2.legend_elements(),loc="upper left", title="Classes")

plt.show()
```

```python
plt.figure(figsize=(15,8))
plt.suptitle('Distribution',color='blue',fontsize=15)
plt.subplots_adjust(hspace=.5, top=.93)

for i in range(1,14):
    plt.subplot(3,5,i)
    plt.hist(X.iloc[:,i-1])
    plt.title(X.columns[i-1],color='green')
    l=f'Skewness: {round(X.iloc[:,i-1].skew(),2)}\nKurtosis: {round(X.iloc[:,i-1].kurtosis(),2)}'
    plt.xlabel(l,color='red')

plt.show()
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
X_scl=pd.DataFrame(sc.fit_transform(X))
```

```python
plt.figure(figsize=(15,8))
plt.suptitle('Distribution after Standardisation',color='blue',fontsize=15)
plt.subplots_adjust(hspace=.5, top=.93)

for i in range(1,14):
    plt.subplot(3,5,i)
    plt.hist(X_scl.iloc[:,i-1])
    plt.title(X.columns[i-1],color='green')
    l=f'Skewness: {round(X.iloc[:,i-1].skew(),2)}\nKurtosis: {round(X.iloc[:,i-1].kurtosis(),2)}'
    plt.xlabel(l,color='red')

plt.show()
```

```python
# Box Plot
plt.figure(figsize=(20,25))
plt.subplots_adjust(hspace=.5, top=0.96)

fly={}
l=[]
for i in range(1,14):
    plt.subplot(5,6,i)
    plt.title(X.columns[i-1],color='green')
    plt.suptitle("Outlier", fontsize=15,color='blue')
    box=plt.boxplot(X.iloc[:,i-1],showfliers=True,vert=False)
    if len([item.get_ydata() for item in box['fliers']][0])>0:
        fly[X.columns[i-1]]=[item.get_ydata() for item in box['fliers']]
```

# Loading Data

```python
import pandas as pd

data=pd.read_csv('train.csv',index_col=0)
data.head()

print(f'Shape of the Data {data.shape}')
print(f'No. of Columns: {data.shape[1]}')
print(f'No. of Rows: {data.shape[0]}')
print(f'Size of data: {data.size}')
```

```python
data.info(memory_usage=True)
```

# EDA

```python
data.describe()
```

```python
print('Columns:')
print(data.columns)
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
# Data Distribution (Univariate Analysis)
plt.figure(figsize=(8, 20))

for i, j in enumerate(data.columns):

    if data[j].dtype == 'O':
        plt.subplot(20, 3, i + 1)
        plt.title(j)
        plt.bar(data[j].value_counts().keys(), data[j].value_counts().values)
    else:
        plt.subplot(20, 3, i + 1)
        plt.title(j)
        data[j].hist()

plt.tight_layout()
plt.show()
```

```python
data.skew()
```

```python
data.kurt()
```

```python
out={}
plt.figure(figsize=(20,20))
x=1

for i,j in enumerate(data.columns):
  if data[j].dtypes!='O':
    plt.subplot(10,2,x)
    box=plt.boxplot(data[j],vert=False)
    outlier=[item.get_ydata() for item in box['fliers']]
    out[j]=outlier
    plt.title(j)
    x+=1

plt.tight_layout()
# plt.show()
```

```python
from sklearn.model_selection import train_test_split

X,Y=data.drop('yield',axis=1),data['yield']
```

```python
from sklearn.ensemble import IsolationForest

iso=IsolationForest()
values=iso.fit_predict(X,Y)

index=[]
for i,j in enumerate(values):
    if j==-1:
        index.append(i)


data.loc[index]
```

# Model Building

```python
from xgboost import XGBRegressor
from sklearn.metrics import mean_absolute_error


x_train,x_test,y_train,y_test=train_test_split(X,Y,random_state=2)

xgb=XGBRegressor()

xgb.fit(x_train,y_train)

pred=xgb.predict(x_test)

print(f'MAE: {mean_absolute_error(y_test,pred)}')
```

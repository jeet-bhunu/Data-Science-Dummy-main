```python
import statsmodels.api as sm
from patsy import dmatrices
```

```python
df = sm.datasets.get_rdataset("Guerry", "HistData").data
```

```python
df.head()
```

```python
df.columns
```

```python
df.isnull().sum()
```

```python
dfc=df.dropna()
```

```python
y, X = dmatrices('Lottery ~ Literacy + Wealth + Region', data=dfc, return_type='dataframe')
```

```python
mod = sm.OLS(y, X)
```

```python
res = mod.fit()
```

```python
print(res.summary())
```

## Correlation

```python
corr_col=[]
no_corr=[]

for i in data.columns[:-1]:
    # Calculate the point-biserial correlation between a continuous feature (col1) and a binary target variable (target)
    corr, p_value = pointbiserialr(data[i], data['target'])
    if p_value<0.05:
        corr_col.append(i)
    else:
        no_corr.append(i)
    # Print the correlation coefficient and the p-value
    print(i)
    print('Point-Biserial Correlation Coefficient:', round(corr,2))
    print('P-Value:', p_value)
    print('----------------')

# Create a bar chart to show the correlation status of each column
fig, ax = plt.subplots(figsize=(8,5))
ax.bar(['Columns with correlation', 'Columns with no correlation'], [len(corr_col), len(no_corr)], color=['#3cbe74', '#f15a29'], alpha=0.8)
ax.set_title('Correlation Status of Columns', fontsize=18, fontweight='bold')
ax.set_ylabel('Number of columns', fontsize=14)
ax.tick_params(axis='x', which='both', length=0, labelsize=12)
ax.tick_params(axis='y', which='major', labelsize=12)
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.text(0.73, 0.8, f'{len(corr_col)} columns with correlation:\n{", ".join(corr_col)}', transform=ax.transAxes, ha='center', va='center', fontsize=14)
ax.text(0.73, 0.25, f'{len(no_corr)} columns with no correlation:\n{", ".join(no_corr)}', transform=ax.transAxes, ha='center', va='center', fontsize=14)
plt.show()
```

## PCA

```python
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
import numpy as np

pca = PCA(n_components=len(x_scl.columns))
pca.fit(x_scl)

# Calculate the explained variance ratio for each principal component
explained_variance_ratio = pca.explained_variance_ratio_

# Calculate the cumulative explained variance ratio
cumulative_variance_ratio = np.cumsum(explained_variance_ratio)

# Plot the scree plot
plt.plot(range(1, len(cumulative_variance_ratio) + 1), cumulative_variance_ratio, marker='o')
plt.xlabel('Number of Principal Components')
plt.ylabel('Cumulative Explained Variance Ratio')
plt.title('Scree Plot')

# Add a horizontal line at the 0.95 variance explained threshold
plt.axhline(y=0.95, color='r', linestyle='-')
plt.text(0.5, 0.9, '95% threshold', color='red', fontsize=10, transform=plt.gca().transAxes)

plt.show()
```

Principal component analysis, or PCA, is a statistical procedure that allows you to summarize the information content in large data tables by means of a smaller set of [“summary indices”](#summary_indexing)
 that can be more easily visualized and analyzed. The underlying data can be measurements describing properties of production samples, chemical compounds or reactions, process time points of a continuous process, batches from a batch process, biological individuals or trials of a DOE-protocol, for example.
 
Principal component analysis today is one of the most popular multivariate statistical techniques. It has been widely used in the areas of pattern recognition and signal processing and is a statistical method under the broad title of <a id='factor'></a>[“factor analysis”](#factor1).


<h3>PCA is the mother method for MVDA</h3>

<p>PCA forms the basis of <strong>multivariate data analysis</strong> based on projection methods. The most important use of PCA is to represent a multivariate data table as smaller set of variables (summary indices) in order to observe trends, jumps, clusters and outliers. This overview may uncover the relationships between observations and variables, and among the variables.</p>

<p>PCA goes back to Cauchy but was first formulated in statistics by <a href='https://www.britannica.com/biography/Karl-Pearson'>Pearson</a>, who described the analysis as finding <a href='https://sci-hub.ru/10.1080/14786440109462720'>“<em>lines and planes of closest fit to systems of points in space”</em></a>[<a href="https://pdfs.semanticscholar.org/70a9/769e8027b48bce5a9ce3a82d0ce88e0f0730.pdf">Jackson, 1991</a>].</p>

<p>PCA is a very flexible tool and allows analysis of datasets that may contain, for example, multicollinearity, missing values, categorical data, and imprecise measurements. The goal is to extract the important information from the data and to express this information as a set of summary indices called <strong>principal components</strong>.</p>


<p>Statistically, PCA finds lines, planes and hyper-planes in the K-dimensional space that approximate the data as well as possible in the least squares sense. A line or plane that is the least squares approximation of a set of data points makes the variance of the coordinates on the line or plane as large as possible.</p>

<img src='https://www.sartorius.com/resource/image/545520/16x9/1050/590/1c5efa21be55247833ee4b2efc948bf7/93D936097F494FF8B646D1309A714921/blog-photo-b24-2-figure-3.jpg'>
<em>PCA creates a visualization of data that minimizes residual variance in the least squares sense and maximizes the variance of the projection coordinates.</em>

<h3>How PCA works</h3>

<p>Consider a matrix X with N rows (aka "observations") and K columns (aka "variables"). For this matrix, we construct a variable space with as many dimensions as there are variables (see figure below). Each variable represents one coordinate axis. For each variable, the length has been standardized according to a scaling criterion, normally by scaling to unit variance.</p>

<img src='https://www.sartorius.com/resource/image/545524/16x9/1050/590/84ec84f49de3f1756ecdaf5c27be8b4/036D82B5AD533FC92C80102BE6AA83EB/blog-photo-b24-3-unit-variance-en-b-00367-sartorius.jpg'>

<p>In the next step, each observation (row) of the X-matrix is placed in the K-dimensional variable space. Consequently, the rows in the data table form a swarm of points in this space.</p>

<img src='https://www.sartorius.com/resource/image/545526/16x9/1050/590/c4a6ad647ef3ad2bc5195a90d1d3378f/9B18DA18F3C200EDA6B206388E586F49/blog-photo-b24-4-figure-3-en-b-00368-sartorius.jpg'>

<em>The observations (rows) in the data matrix X can be understood as a swarm of points in the variable space (K-space).</em>

<h3>Mean centering</h3>
<p>Next, mean-centering involves the subtraction of the variable averages from the data. The vector of averages corresponds to a point in the K-space.</p>



<em>In the mean-centering procedure, you first compute the variable averages. This vector of averages is interpretable as a point (here in red) in space. The point is situated in the middle of the point swarm (at the center of gravity).</em>

```python
from sklearn.decomposition import PCA
from sklearn.datasets import load_iris

import matplotlib.pyplot as plt
import pandas as pd
```

```python
data=load_iris().data
data.shape
```

```python
pca=PCA(n_components = 2)
```

```python
trans=pd.DataFrame(pca.fit_transform(data),columns=['PC1','PC2'])
```

```python
pca.explained_variance_ratio_
```

```python
%matplotlib inline
scatter=plt.scatter(x='PC1',y='PC2',data=trans,c=load_iris().target)
legend1=plt.legend(*scatter.legend_elements(),
                    loc="lower right", title="Classes")
plt.show()
```

```python
scatter=plt.scatter(range(150),PCA(1).fit_transform(data),c=load_iris().target)
legend1 = plt.legend(*scatter.legend_elements(),
                    loc="upper left", title="Classes")
# plt.add_artist(legend1)
plt.show()
```

<span><strong>Factor Analysis</strong></span>

<p>Factor analysis is a technique in mathematics that we use to reduce a larger number into a smaller number.</p>

<p>Moreover, it is a part of General Linear Model (GLM) and it believes several theories that contain no multicollinearity, linear relationship, true <a href="https://www.toppr.com/guides/fundamentals-of-business-mathematics-and-statistics/correlation-and-regression-cma/">correlation</a>, and relevant variables into the analysis among factors and variables.</p>

<strong>Types of Factor Analysis</strong>

<strong>1. Principal component analysis</strong>
<p>It is the most common method which the researchers use. Also, it extracts the maximum variance and put them into the first factor. Subsequently, it removes the variance explained by the first factor and extracts the second factor. Moreover, it goes on until the last factor.</p>

<strong>2. Common Factor Analysis</strong>

<p>It’s the second most favoured technique by researchers. Also, it extracts common variance and put them into <a href="https://www.toppr.com/guides/maths/be-my-multiple-ill-be-your-factor/factors-of-a-number/">factors</a>. Furthermore, this technique doesn’t include the variance of all variables and is used in SEM.</p>


<strong>3. Image Factoring</strong>

<p>It is on the basis of the correlation matrix and makes use of OLS regression technique in order to predict the factor in image factoring.</p>

<strong>4. Maximum likelihood method</strong>

<p>It also works on the correlation matrix but uses a maximum likelihood method to factor.</p>



<h4><span><strong>5. Other methods of factor analysis</strong></span></h4>
<p>Alfa factoring outweighs least squares. Weight square is another regression-based method that we use for factoring.</p>

<p><strong>Factor loading-</strong> Basically it the correlation coefficient for the factors and variables. Also, it explains the variable on a particular factor shown by variance.</p>

<p><strong>Eigenvalues-</strong> Characteristics roots are its other name. Moreover, it explains the variance shown by that particular factor out of the total variance. Furthermore, commonality column helps to know how much variance the first factor explained out of total variance.</p>

<p><strong>Factor Score-</strong> It’s another name is the component score. Besides, it’s the score of all rows and columns that we can use as an index for all variables and for further analysis. Moreover, we can standardize it by multiplying it with a common term.</p>

<p><strong>Rotation method-</strong> This method makes it more reliable to understand the output. Also, it affects the eigenvalues method but the eigenvalues method doesn’t affect it. Besides, there are 5 rotation methods: (1) No Rotation Method, (2) Varimax Rotation Method, (3) Quartimax Rotation Method, (4) Direct Oblimin Rotation Method, and (5) Promax Rotation Method.</p>

<h3><span><strong>Assumptions of Factor Analysis</strong></span></h3>

<p>Factor analysis has several assumptions. These include:</p>

<ol><li>There are no outliers in the data.</li>
<li>The sample size is supposed to be greater than the factor.</li>
<li>It is an interdependency method so there should be no perfect multicollinearity between the variables.</li>
<li>Factor analysis is a linear function thus it doesn’t require homoscedasticity between variables.</li>
<li>It is also based on the linearity assumption. So, we can also use non-linear variables. However, after a transfer, they change into a linear variable.</li>
<li>Moreover, it assumes interval data.</li>
</ol>

<h3><span><strong>Key Concepts of Factor Analysis</strong></span></h3>

<p>It includes the following key concept:</p>

<p>Exploratory factor analysis- It assumes that any variable or indicator can be associated with any factor. Moreover, it is the most common method used by researchers. Furthermore, it isn’t based on any prior theory.</p>

<p>Confirmatory Factor Analysis- It is used to determine the factors loading and factors of measured variables, and to confirm what it expects on the basis of pre-established assumption. Besides, it uses two approaches:</p>
<ol><li>The Traditional Method</li>
<li>The SEM Approach</li>
</ol>

<h2>Summary Indices</h2>
<div style='padding:5px;'>
<img src='https://www.sp6.io/wp-content/uploads/2020/08/What-is-Summary-Indexing.jpg' width="300" height="300">
</div>

Many people have heard of summary indexing, yet have not made use of them.  One reason for this is that you may not realize you need them until you need them.  Summary indexes, as the name implies, allows for the storage of summarized data over time.  A good use case for a summary index is queries that require the summarizing/trending of large amounts of data over a longer period of time. 

For example, with the onset of increased teleworking, my customer had an increased need to monitor the activity and resources used around teleworking (VPN activity, login/logoff, concurrent users, zoom activity, etc.). I was required to track hourly metrics over a longer period of time, such as hourly metrics over a period of weeks.  When using good spl structure and tuning, the query still failed to return results in a reasonable amount of time.  This type of query is a great candidate for summary indexing.   

There are two parts to this:  
1). define a query to populate a summary index using smaller chunks of data, and 

2). Define a query to run against the summarized index to calc your final results.

The summary indexing solution allows us to take these bite-size calculations of our data, and store those results in a separate index.  The smaller amount of data has not only gotten us a head start in our calculations, but has also allowed for a smaller amount of data for us to query through.

**Noisy dataset**

```python
# Summary of the columns.
def columns_summary(df):
    cs={}
    for i in df.columns:
        cs[i]=df[i].describe()
        cs[f'null {i}']=df[i].isnull().sum()
        cs[f'null % {i}']=round((df[i].isnull().sum()/len(df[i].index))*100,2)
    return cs

summary=columns_summary(data)

summary_df=pd.DataFrame(summary).fillna('🥊')
summary_df
```

```python
# Total Fighters in dataset


r_fighters=list(data.R_fighter.value_counts().keys())
b_fighters=list(data.B_fighter.value_counts().keys())
unique_fighters=set(r_fighters+b_fighters)
combined = pd.concat([data['R_fighter'], data['B_fighter']])

print(f'Total Fighters in dataset: {len(combined)}')
print(f'Total Unique fighters in the dataset {len(unique_fighters)}')

# for i in range()
# plt.figure(figsize=(10,25))
combined.value_counts()[:20].plot(kind='bar',color='green',figsize=(10,5))

plt.title('Bar Chart of Top 20 Most Common Fighter Names in the Dataset')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# get the value counts for the Winner column
winner_counts = data.Winner.value_counts()

# create a dictionary to map winner names to colors
color_map = {'Red': 'red', 'Blue': 'blue', 'Draw': 'yellow'}

plt.figure(figsize=(7,4))
# create a bar plot with custom colors
ax = winner_counts.plot(kind='bar', color=[color_map[winner] for winner in winner_counts.index])

# add values on top of the bars
for i, v in winner_counts.reset_index().iterrows():
    ax.text(i, v.Winner + 0.9, str(v.Winner), color='purple', ha='center')

# add a legend for each color
legend_handles = []
for winner, color in color_map.items():
    legend_handles.append(plt.Rectangle((0,0),1,1,fc=color, edgecolor = 'green'))
ax.legend(legend_handles, color_map.keys())

# show the plot
plt.show()
```

```python
# Probability Plot
# get the value counts for the Winner column
winner_counts = data.Winner.value_counts()

# create a bar plot with custom colors and add title
ax = (winner_counts/len(data)).plot(kind='bar', color=['#FF4136', '#0074D9', '#FFDC00'])
plt.title('Winning Probability by Fighter Color', fontsize=18, color='#333333')

# set x-axis and y-axis labels
ax.set_xlabel('Fighter Color', fontsize=14, color='#333333')
ax.set_ylabel('Probability', fontsize=14, color='#333333')

# set x-tick labels
ax.set_xticklabels(['Red', 'Blue', 'Draw'], rotation=0, fontsize=12, color='#333333')

# add values on top of bars
for p in ax.patches:
    ax.annotate(f'{p.get_height():.2f}', (p.get_x() + p.get_width() / 2., p.get_height()), 
                ha='center', va='center', fontsize=12, color='#FFFFFF', xytext=(0, 10), 
                textcoords='offset points')

# remove borders and ticks from plot
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['bottom'].set_color('#CCCCCC')
ax.spines['left'].set_color('#CCCCCC')
ax.tick_params(axis='x', colors='#CCCCCC')
ax.tick_params(axis='y', colors='#CCCCCC')

# set background color
ax.set_facecolor('#F7F7F7')

# show the plot
plt.show()
```

```python
# Top winner
l=[]
for i,j in enumerate(data.Winner):
    if j =="Red":
        l.append(data.R_fighter.loc[i])
    elif j=="Blue":
        l.append(data.B_fighter.loc[i])
#         print(i)
    

winners=pd.Series(l)
print(winners.value_counts())
```

```python
# Bar Chart for top 10 winners
# Get the value counts for the winners column
winner_counts = winners.value_counts()

# Get the top 10 winners
top_winners = winner_counts[:10]

# Create a horizontal bar plot with custom colors and add title
fig, ax = plt.subplots(figsize=(8, 5))
ax.barh(top_winners.index, top_winners.values, color='#480607')

# Add the total count on top of each bar
for i, v in enumerate(top_winners.values):
    ax.text(v +.5, i + .15, str(v), color='#ff2052', fontweight='bold')

# Set x-axis and y-axis labels
ax.set_xlabel('Number of Wins', fontsize=14, color='#ff2052')
ax.set_ylabel('Fighter', fontsize=14, color='#ff2052')

# Set title and adjust layout
ax.set_title('Top 10 Winners', fontsize=18, color='#ff2052')
fig.tight_layout()

# Remove borders and ticks from plot
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['bottom'].set_color('#5d8aa8')
ax.spines['left'].set_color('#5d8aa8')
ax.tick_params(axis='x', colors='#e52b50')
ax.tick_params(axis='y', colors='#9966cc')

# Set background color
ax.set_facecolor('#F7F7F7')

# Show the plot
plt.show()
```

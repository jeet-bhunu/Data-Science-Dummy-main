```python
data1=data.drop(['dismissal_kind','player_dismissed','fielder','extras_type'],axis=1)
data1.head()
```

```python
data1.index.value_counts()
```

```python
%%time
import matplotlib.pyplot as plt

c = 1
for i in data1.batsman.value_counts().keys()[:50]:
    plt.figure(figsize=(25,55))
    plt.subplot(300, 3, c)
    data1[data['batsman'] == i].batting_team.value_counts().plot(kind='bar')
    plt.title(i)
    c += 1
    plt.show()
# plt.tight_layout()
#     print(i)
```

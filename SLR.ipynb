```python
import pandas as pd

data=pd.read_csv('https://raw.githubusercontent.com/slmsshk/DataSet-main/main/NewspaperData.csv')

data

# 
data.describe(include='all').fillna('😏')

# scatter plot, corr, heatmaps
import seaborn as sns
import matplotlib.pyplot as plt
import warnings

warnings.filterwarnings('ignore')
corr = data.corr()
sns.heatmap(corr,annot=True)
plt.show()

plt.title('Sunday~Daily Relation')
plt.scatter(data.daily,data.sunday)
plt.show()

X,Y=data[['daily']],data[['sunday']]

# arrays, Series, DataFrame
X,Y

from sklearn.linear_model import LinearRegression

lr = LinearRegression()

lr.fit(X,Y)

slope

intercept = lr.intercept_

slope = lr.coef_

# slope(x1)+intercept

new = 200

predicted= (slope*new) +intercept
predicted

lr.predict([[200]])

lr.score(X,Y)
```

```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(42)
num_points = 100
x = np.linspace(0, 10, num_points)
y = 5 - 0.7 * x + np.random.normal(0, 2, num_points)

plt.scatter(x, y, label="Negative Correlation", color='b', alpha=0.6)

z = np.polyfit(x, y, 1)
p = np.poly1d(z)
plt.plot(x, p(x), "r--", label="Trendline")

plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Negative Correlation Example")

plt.legend()

plt.grid(True)
plt.show()
```

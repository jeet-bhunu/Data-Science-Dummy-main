```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt


import warnings
warnings.filterwarnings('ignore')
```

```python
df=pd.read_csv(r"\Users\Slmss\Documents\GitHub\DataSet\brain_networks.csv",header=None).dropna()
```

```python
df.head()
```

```python
df1=df.drop(0,axis=1)
df1.head()
```

```python
from sklearn.preprocessing import StandardScaler ,LabelEncoder
```

```python
sc=StandardScaler()
le=LabelEncoder()
```

```python
df1.iloc[2,:]=le.fit_transform(df1.iloc[2,:])
df1.head()
```

```python
df2=df1.copy()
df2.iloc[:,:]=sc.fit_transform(df1)
df2.head()
```

```python
from sklearn.manifold import TSNE
```

```python
tsne=TSNE()
```

```python
score=pd.DataFrame(tsne.fit_transform(df2),columns=['A','B'])
```

```python
plt.figure(figsize=(15,5))
plt.scatter(score.A,score.B)
# plt.legend()
plt.show()
```

```python
from sklearn.cluster import KMeans
plt.figure(figsize=(25,20))

l=[2,17,4]
n=['hemisphere','Network','Node']
c=1
for i in l:
    k=KMeans(i)
    plt.subplot(3,1,c)
    scatter=plt.scatter(score.A,score.B,c=k.fit_predict(df2))
    legend1 = plt.legend(*scatter.legend_elements(),
                    loc="lower left", title=n[c-1])
    plt.title(n[c-1])
    c+=1
# plt.legend()
# plt.show()
```

```python
df_network=df.T
df_network.columns=df.T.loc[0]
```

```python
dfn=df_network.drop(0,axis=0)
dfn['network']=dfn['network'].astype(int)
dfn1=dfn.sort_values(by='network').reset_index()
```

```python
dfn.head()
```

```python
plt.figure(figsize=(20,30))

for i in range(len(dfn.network.value_counts())):
    tsne=TSNE()
    x=dfn[dfn.network==i+1]
    score=pd.DataFrame(tsne.fit_transform(x.drop(['network','node','hemi'],axis=1).T),columns=['a','b'])
    plt.subplot(6,3,i+1)
    plt.scatter(score.a,score.b,label='network '+ str(i),alpha=0.5)
    plt.legend()
```

```python
plt.figure(figsize=(10,10))

for i in range(len(dfn.node.value_counts())):
    tsne=TSNE()
    x=dfn[dfn.node==str(i+1)]
    score=pd.DataFrame(tsne.fit_transform(x.drop(['network','node','hemi'],axis=1).T),columns=['a','b'])
    plt.subplot(2,2,i+1)
    plt.scatter(score.a,score.b,label='Node '+ str(i+1),alpha=0.5)
    plt.legend()
```

```python
u=dfn.hemi.unique()
dfn.hemi=le.fit_transform(dfn.hemi)

plt.figure(figsize=(15,5))
for i in range(len(dfn.hemi.value_counts())):
    tsne=TSNE()
    x=dfn[dfn.hemi==i]
    score=pd.DataFrame(tsne.fit_transform(x.drop(['network','node','hemi'],axis=1).T),columns=['a','b'])
    plt.subplot(1,3,i+1)
    plt.scatter(score.a,score.b,label='Hemisphere '+ u[i],alpha=0.8)
    plt.legend()
```

```python
import pandas as pd
import matplotlib.pyplot as plt


import seaborn as sns
```

```python
data=pd.read_csv('train.csv',index_col=0)
data.head(2)
```

```python
corr=data.corr()

sns.heatmap(corr, annot=True, vmin=-1, vmax=1, cmap='coolwarm', square=True)
plt.show()
```

```python
data.describe(include='all').fillna('-.-')
```

```python
data.clarity.value_counts().keys()
```

```python
# Categorical Data Distribution
plt.figure(figsize=(10,10),dpi=512)
plt.subplot(2,2,1)
colors = sns.color_palette('Set2', n_colors=5)
ax = data.cut.value_counts().plot(kind='bar', color=colors)
for p in ax.patches:
    ax.text(p.get_x() + p.get_width()/2, p.get_height(), int(p.get_height()), ha='center', va='bottom')
labels = ['Ideal', 'Premium', 'Very Good', 'Good', 'Fair']
ax.set_xticklabels(labels, rotation=45)
plt.title('Cut Distribution')

plt.subplot(2,2,2)
colors = sns.color_palette('Set1', n_colors=5)
ax = data.color.value_counts().plot(kind='bar', color=colors)
for p in ax.patches:
    ax.text(p.get_x() + p.get_width()/2, p.get_height(), int(p.get_height()), ha='center', va='bottom')
labels = ['G', 'E', 'F', 'H', 'D', 'I', 'J']
ax.set_xticklabels(labels, rotation=45)
plt.title('Color Distribution')

plt.subplot(2,2,3)
colors = sns.color_palette('Set3', n_colors=5)
ax = data.clarity.value_counts().plot(kind='bar', color=colors)
for p in ax.patches:
    ax.text(p.get_x() + p.get_width()/2, p.get_height(), int(p.get_height()), ha='center', va='bottom')
labels = ['SI1', 'VS2', 'VS1', 'SI2', 'VVS2', 'VVS1', 'IF', 'I1']
ax.set_xticklabels(labels, rotation=45)
plt.title('Clarity Distribution')

plt.tight_layout()
plt.show()
```

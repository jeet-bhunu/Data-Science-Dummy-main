```python
import pandas as pd
```

# Train Data

```python
train=pd.read_csv('train.csv',index_col=0)
```

```python
train.head()
```

```python
train.describe(include='all').fillna('-')
```

```python
X,Y=train.iloc[:,:-1],train.quality
```

# Scaling Data

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
Y=le.fit_transform(Y)
```

```python
train.shape
```

```python
from sklearn.preprocessing import StandardScaler
std=StandardScaler()
x_scl=std.fit_transform(X)
```

# DR

## TSNE

```python
from sklearn.manifold import TSNE
```

```python
tsne=TSNE()
final_X=pd.DataFrame(tsne.fit_transform(x_scl),columns=['a','b'])
```

## PCA

```python
from sklearn.decomposition import PCA

pca=PCA()
```

```python
pca.fit(x_scl)
```

```python
import matplotlib.pyplot as plt

plt.plot(pca.explained_variance_ratio_)
```

```python
pcs=PCA(4)
final_X=pd.DataFrame(pcs.fit_transform(x_scl),columns=('a','b','c','d'))
```

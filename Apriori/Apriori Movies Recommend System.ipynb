```python
#import library
import numpy as np
import pandas as pd

from mlxtend.frequent_patterns import apriori,association_rules

import warnings
warnings.filterwarnings('ignore')
```

```python
#read csv file
movie = pd.read_csv("https://raw.githubusercontent.com/slmsshk/DataSet/main/movies_anr.csv",index_col = None)
movie
```

```python
# Fillna with 0 
df1=movie.iloc[:,5:]
df1.head()
```

```python
# Building the model 
movie_apiori = apriori(
df1, min_support = 0.1, use_colnames = True) 
```

```python
# Collecting the inferred rules in a dataframe 
rules = association_rules(movie_apiori, metric ="lift") 
rules = rules.sort_values(['confidence', 'lift'], ascending =[False, False]) 
rules.head(50)
```
